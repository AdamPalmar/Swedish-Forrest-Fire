```python
import numpy as np
import pandas as pd
from pivottablejs import pivot_ui
import matplotlib.pyplot as plt
import matplotlib
from sklearn import preprocessing 
matplotlib.style.use('ggplot')
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 20, 20
rcParams.update({'font.size': 30})
rcParams.update({'figure.autolayout': True})

data = pd.read_csv("school_fire_cases_1998_2014.csv")
```

```python
pivot_ui(data)
```

```python
year_range = range(1998,2014)



mean_list = list()
std_list = list()
var_list = list()

for i in year_range:
    print i
    #Mean of 1998
    print "Mean", mean_list.append(round(data.where(data['Year'] == i)['Cases'].dropna().mean(),2))
    #0.41 fires on average
    #Standard
    print "Standard deviation", std_list.append(round(data.where(data['Year'] == i)['Cases'].dropna().std(),2))
    #Var
    print "Variance",var_list.append(round(data.where(data['Year'] == i)['Cases'].dropna().var(),2))

fire_cases = pd.DataFrame(index = year_range,columns=["mean",'std','var'])

for i,x in enumerate(year_range):
    fire_cases.loc[x] = [mean_list[i],std_list[i],var_list[i]]
```

```python

import numpy as np
from scipy import stats


col_name = 'medianIncome'
income_mean = data_simple[col_name].mean()
income_std = data_simple[col_name].std()
h = data_simple[col_name].values
h = sorted(h)

# Number of samples
N = 290

# Mean
mu = income_mean

# Standard deviation
s = income_std

# Number of bins in histogram
nbins = 40

# Generate samples from the Normal distribution
X = np.mat(np.random.normal(mu,s,N)).T 


# Plot the histogram
f = figure()
f.hold()
title('Normal distribution median Income')
plt.hist(h, bins=nbins, normed=True)

# Over the histogram, plot the theoretical probability distribution function:
x = linspace(X.min(), X.max(), 1000)
pdf = stats.norm.pdf(x,loc=income_mean,scale=income_std)
plt.plot(x,pdf,'-',color='blue')

# Compute empirical mean and standard deviation
mu_ = X.mean()
s_ = X.std(ddof=1)

print "Theoretical mean: ", mu
print "Theoretical std.dev.: ", s
print "Empirical mean: ", mu_
print "Empirical std.dev.: ", s_

plt.savefig('normal_income_red')

plt.show()
```
