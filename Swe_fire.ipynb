```python
import numpy as np
import pandas as pd
from pivottablejs import pivot_ui
import matplotlib.pyplot as plt
import matplotlib
from sklearn import preprocessing 
matplotlib.style.use('ggplot')
%matplotlib inline

data = pd.read_csv("school_fire_cases_1998_2014.csv")
```

```python
pivot_ui(data)
```

```python
year_range = range(1998,2014)



mean_list = list()
std_list = list()
var_list = list()

for i in year_range:
    print i
    #Mean of 1998
    print "Mean", mean_list.append(round(data.where(data['Year'] == i)['Cases'].dropna().mean(),2))
    #0.41 fires on average
    #Standard
    print "Standard deviation", std_list.append(round(data.where(data['Year'] == i)['Cases'].dropna().std(),2))
    #Var
    print "Variance",var_list.append(round(data.where(data['Year'] == i)['Cases'].dropna().var(),2))

fire_cases = pd.DataFrame(index = year_range,columns=["mean",'std','var'])

for i,x in enumerate(year_range):
    fire_cases.loc[x] = [mean_list[i],std_list[i],var_list[i]]
```

```python
print fire_cases.to_latex()
```

```python
fire_cases['mean'].plot()
plt.xlabel('Years')
plt.ylabel('Mean of Fire cases')
plt.savefig('Mean_fires')
plt.show()
```

```python
fire_cases['std'].plot()
plt.xlabel('Years')
plt.ylabel('Std of Fire cases')
plt.savefig('std_fires')
plt.show()
```

```python
fire_cases['var'].plot()
plt.xlabel('Years')
plt.ylabel('Var of Fire cases')
plt.savefig('var_fires')
plt.show()
```
